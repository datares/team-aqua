```python
walmart_df = pd.read_csv('walmart.csv')
walmart_df = walmart_df.drop(['Crawl Timestamp', 'Product Url', 'Item Number', 'Gtin', 'Package Size', 'Postal Code'], axis =1)
walmart_df.head()
```

```python
fig = plt.figure()
#plt.figure(figsize=(8,4))
ax = sns.countplot(x = 'Product Category', palette = 'Spectral', data = df2,order = df2['Product Category'].value_counts().index )
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha = 'right')
fig.savefig("categories.png", bbox_inches='tight', dpi=600)
```

```python
walmart_df['Category'] = walmart_df['Category'].astype(str)
def new_cat (data):
    if '|' in str(data):
        return data.split('|')[0]
    return 'NA'


walmart_df['Category'] = walmart_df['Category'].map(new_cat)
```

```python
walmart_df=walmart_df.dropna(axis=0)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'Feature'], inplace = True)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'Jewelry & Watches'], inplace = True)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'Seasonal'], inplace = True)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'NA'], inplace = True)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'Cell Phones'], inplace = True)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'Music'], inplace = True)
walmart_df.drop(walmart_df.index[walmart_df['Category'] == 'Auto & Tires'], inplace = True)
```
